```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('Mall_Customers.csv')
```

```python
dataset
```

```python
#iloc-> instead of defining input one by one we can use iloc to get all the column name (row wise:,column wise:)
#illoc ->[:,3:5] [row take all : column 3th and 5th column]
```

```python
X = dataset.iloc[:, 3:4].values
```

```python
X
```

```python
 X = dataset.iloc[:, 3:5].values
```

```python
X
```

```python
from sklearn.cluster import KMeans
list1 = []
for i in range(1, 11):   #1 to 11 loop
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42) #pass i value 1 to 11 (we can define 12  or more than that also)
    kmeans.fit(X)
    list1.append(kmeans.inertia_) #inertia method every time execute store value in list
plt.plot(range(1, 11), list1)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# 2 point we have targeting ->No of clusters-> 3 and 5 (which one is closer 5 that is good targeting)
```

```python
# this is used to find how many cluster forming from the given input
```

```python
list1
```

```python
 #5 (which one is closer 5 that is good targeting , so we pass to the cluster)
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters = 5, init = 'k-means++', random_state = 42) #5 is good target so define only 5
y_kmeans = kmeans.fit_predict(X) #model also created and prediction also created
```
