```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('Mall_Customers.csv')
```

```python
dataset
```

```python
 X = dataset.iloc[:, 3:5].values
```

```python
X
```

```python

from sklearn.cluster import OPTICS

# Fit OPTICS
optics = OPTICS(min_samples=5)
y_pred = optics.fit_predict(X)

```

```python
y_pred
```

```python
# Plot
plt.scatter(X[:, 0], X[:, 1], c=y_pred, cmap='viridis', s=50)
plt.title(f"OPTICS Clustering\nEstimated clusters: {len(np.unique(y_pred)) - (1 if -1 in y_pred else 0)} (Noise: {-1 in y_pred})")
plt.show()
```

```python
supervised=dataset
```

```python
supervised
```

```python
supervised['Cluster_group']=y_pred
```

```python
supervised
```

```python
supervised.to_csv("clusterOpticsalgorithm.csv",index=False) #we need to save becoz we added new column
```
